#**결측치 처리**

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'a' : [1, 1, 3, 4, 5], 'b' : [2, 3, np.nan, 3, np.nan], 'c' : [3, 4, np.nan, 6, 4]})
```

```python
df
```

```python
#곃측치유무 판별
df.isnull()
```

```python
#결측치 개수 확인
df.isnull().sum()
```

```python
#결측값 포함된 행 삭제
df.dropna()
```

```python
df
```

```python
#결측값 포함된 열 삭제
df.dropna(axis=1)
```

```python
df
```

```python
#결측값 특정 값으로 대체하기
df.fillna(0)
```

```python
#결측값 앞의 값으로 대체하기
df.fillna(method='ffill') 
```

```python
#결측값 뒤의 값으로 대체하기
df.fillna(method='bfill') 
```

```python
df = pd.DataFrame({'a' : [np.nan, 1, 3, 4, 5], 'b' : [2, 3, np.nan, np.nan, 3], 'c' : [3, 4, np.nan, 6, 4]})
```

```python
df
```

```python
#결측값 변경 횟수 제한
df.fillna(method='ffill', limit=1) 
```

```python
#결측치 평균 값으로 대체
#1. a열 평균 구하기
df.mean()['a']
#2. a열 평균 대체
df.fillna(df.mean()['a'])
```

```python
df
```

```python
#각 열의 평균값으로 대체
print(df.mean())
df.fillna(df.mean())
```

```python
#해당열만 평균값으로 대체
df.fillna(df.mean()[['b','c']])
```

#apply, map을 사용한 데이터 변환

####기본 데이터 변환

```python
import pandas as pd

df = pd.DataFrame({'a' : [1, 2, 3, 4, 5]})
```

```python
#a 가 2 보다 작으면 '2 미만', 4 보다 작으면 '4 미만', 4 보다 크면 '4 이상' 이 저장된 b 칼럼을 추가하기. default는 0이다.
```

```python
#b 열을 0으로 초기화
df['b'] = 0
```

```python
df
```

```python
num1 = df[(df['a']<2)]
num2 = df[(df['a']>=2) & (df['a']<4)]
num3 = df[(df['a']>=4)]
```

```python
df['b'][num1.index] = '2 미만'
df['b'][num2.index] = '4 미만'
df['b'][num3.index] = '4 이상'
```

```python
df
```

####apply 함수를 이용한 데이터 변환

```python
def case_1(x):
  if x<2:
    return '2 미만'
  elif x<4:
    return '4 미만'
  else:
    return '4 이상'
```

```python
df
```

```python
#a가 1이면 'one', 2이면 'two', 3이면 'three', 4이면 'four', 5이면 'five'를 저장하는 d 만들기
```

```python
def case_2(x):
  if x==1:
    return 'one'
  elif x==2:
    return 'two'
  elif x==3:
    return 'three'
  elif x==4:
    return 'four'
  elif x==5:
    return 'five'
```

```python
df['d'] = df['a'].apply(case_2)
```

```python
df
```

```python
df['e'] = df['a'].map(a)
```
