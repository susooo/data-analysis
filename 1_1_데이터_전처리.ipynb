#**조건에 맞는 데이터 추출**

```python
import pandas as pd
df = pd.DataFrame({'a' : [i for i in range(1, 11)], 'b' : [i for i in range(11, 21)], 'c' : [i for i in range(21, 31)]})
```

```python
df
```

```python
#a 가 3 이상인 데이터
df[df['a'] >= 3]
```

```python
#a 가 3 이상인 데이터 중 a, c 열만
df[df['a'] >= 3][['a', 'c']]
```

```python
#a 가 3 이상이고, b 가 16 미만인 데이터
df[(df['a'] >= 3) & (df['b'] < 16)]
```

```python
a = (df['a'] >= 3) & (df['b'] < 16)
df[a]
```

```python
#a 가 3 이하 이거나 7 이상인 데이터
df[(df['a'] <= 3) | (df['a'] >= 7)]
```

```python
#a 가 3 이상이고, b 가 16 미만이거나 c 가 30 인 데이터
df[(df['a'] >= 3) & ((df['b'] < 16) | (df['c'] == 30))]
```
