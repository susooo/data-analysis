#산점도 그리기

####scatterplot

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize':(10, 5)})
import pandas as pd
```

```python
df = sns.load_dataset('iris')
```

```python
#sepal : 꽃잎
#petal : 꽃받침
df.head()
```

```python
df.isnull().sum()
```

```python
sns.scatterplot(data = df, x = 'sepal_width', y = 'sepal_length')
```

```python
sns.scatterplot(data = df, x = 'sepal_width', y = 'sepal_length', hue = 'species')
```

```python
#칼럼안 유일한 값 찾기
df['petal_length'].unique()
```

```python
#petal_length의 값을 3미만은 small, 5미만은 medium, 5이상dmf large로 분류하기
def function1(x):
    if x < 3:
        return 's'
    elif x < 5:
        return 'm'
    else:
        return 'l'
```

```python
df['petal_length2'] = df['petal_length'].apply(function1)
```

```python
df.head(10)
```

```python
#점의 모양으로 분류
sns.scatterplot(data = df, x = 'sepal_width', y = 'sepal_length', hue = 'species', style = 'petal_length2')
```

```python
def function2(x):
    if x < 3:
        return 3
    elif x < 5:
        return 5
    else:
        return 7
```

```python
df['petal_length3'] = df['petal_length'].apply(function2)
```

```python
df.head()
```

```python
#점의 크기로 분류
sns.scatterplot(data = df, x = 'sepal_width', y = 'sepal_length', hue = 'species', size = 'petal_length3')
```

```python
#카테고리별로 따로 그릴 수 있다.
sns.relplot(data = df, x = 'sepal_width', y = 'sepal_length', col = 'species')
```

```python
#회귀선을 그릴 수 있다.
sns.lmplot(data = df, x = 'sepal_width', y = 'sepal_length', hue = 'species')
```

#선도표 그리기

####lineplot

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize':(10, 5)})
import pandas as pd
```

```python
df = sns.load_dataset('flights')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#년도별 승객수 시각화
#선 주위의 하늘색 범위는 12개월치 데이터를 표현한다.
sns.lineplot(data = df, x = 'year', y = 'passengers')
```

```python
#x축에 전체년도 표시
#세미클론 넣으면 출력되는 텍스트 생략된다.
sns.lineplot(data = df, x = 'year', y = 'passengers')
plt.xticks(df['year']);
```

```python
#x축 기울기 추가
sns.lineplot(data = df, x = 'year', y = 'passengers')
plt.xticks(df['year'], rotation = 45);
```

```python
#월별로 그리기
sns.lineplot(data = df, x = 'year', y = 'passengers', hue = 'month')
```

```python
#차트 색상 변경
sns.lineplot(data = df, x = 'year', y = 'passengers', hue = 'month', palette = 'Set2')
```

```python
#점선으로 표현방법 변경
sns.lineplot(data = df, x = 'year', y = 'passengers', hue = 'month', palette = 'Set2', style = 'month')
```

#히트맵 그리기

####heatmap

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize':(10, 5)})
import pandas as pd
```

```python
df = sns.load_dataset('flights')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
#년도별, 월별 승객수를 피벗 테이블로 만든다.
pivot_data = df.pivot("month", "year", "passengers")
```

```python
pivot_data
```

```python
#히트맵 그리기
sns.heatmap(pivot_data)
```

```python
#구분선 추가
sns.heatmap(pivot_data, linewidths=.5)
```

```python
#색상 변경
sns.heatmap(pivot_data, cmap="Blues")
```

```python
#수치입력
sns.heatmap(pivot_data, cmap="Blues", annot = True)
```

```python
#수치 정수로 입력
sns.heatmap(pivot_data, cmap="Blues", annot = True, fmt = 'd')
```

```python
#cbar 가로로 놓기
fig, (ax, cbar_ax) = plt.subplots(2)
```

```python
fig
```

```python
ax
```

```python
cbar_ax
```

```python
#수평적으로 그리기
ax = sns.heatmap(pivot_data, ax=ax,
                  cbar_ax=cbar_ax,
                  cbar_kws={"orientation": "horizontal"})
```

```python
fig
```

```python
#수직적으로 그리기
ax = sns.heatmap(pivot_data, ax=ax,
                  cbar_ax=cbar_ax,
                  cbar_kws={"orientation": "vertical"})
```

```python
fig
```

```python
# cbar를  가로: 세로  85 : 10 크기의 차트와 0.4 간격이 되도록  변경
grid_kws = {"height_ratios": (.85, .1), "hspace": 0.4}

f, (ax, cbar_ax) = plt.subplots(2, gridspec_kw=grid_kws)
ax = sns.heatmap(pivot_data, ax=ax,
                 cbar_ax=cbar_ax,
                 cbar_kws={"orientation": "horizontal"})
```

```python
grid_kws = {"height_ratios": (.85, .1), "hspace": 0.4}

f, (ax, cbar_ax) = plt.subplots(2, gridspec_kw=grid_kws)
ax = sns.heatmap(pivot_data, ax=ax,
                 cbar_ax=cbar_ax,
                 cbar_kws={"orientation": "horizontal"},
                 cmap="Blues", 
                 annot = True, fmt = 'd')
```

#원형차트 그리기

####pie

```python
#한글폰트 사용하기
import matplotlib.pyplot as plt
plt.rcParams['font.family']=['NanumGothic', 'sans-serif']
plt.rcParams['axes.unicode_minus'] = False

import pandas as pd
```

```python
#천만 영화
movie_title = ['명랑', '아바타', '국제시장', '극한직업', '도둑들']
audience = [17613682, 13624328, 14257115, 16264944, 12983330]

data = {'영화제목' : movie_title, '누적관객' : audience}
df = pd.DataFrame(data)
```

```python
#데이터프레임
df
```

```python
#원형차트 그리기
plt.pie(df['누적관객']);
```

```python
#라벨 추가하기
plt.pie(df['누적관객'], labels = df['영화제목']);
```

```python
#비율값 입력하기
plt.pie(df['누적관객'], labels = df['영화제목'], autopct = '%0.2f%%');
```

```python
#colors_list = ['red', 'green', 'blue']
colors_list = ['#A52A2A', '#CD5C5C', '#BC8F8F', '#FFDAB9', '#DEB887']

plt.pie(df['누적관객'], labels = df['영화제목'], autopct = '%0.2f%%', colors = colors_list);
```

```python
#특정부분 강조하기
explode_list = [0, 0.1, 0, 0, 0]

plt.pie(df['누적관객'], labels = df['영화제목'], autopct = '%0.2f%%', colors = colors_list, explode = explode_list);
```
